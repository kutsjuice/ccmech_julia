# Многоточечные связи

Часто при решении задач требуется наложить дополнительные ограничение на решение. В качестве примера таких ограничений можно привести ограничение в результате контакта двух тел или периодические граничные условия. В результате наложения такого рода ограничений перемещения в определенных узлах получаются связаны друг с другом, или говоря иначе – зависят друг от друга. Так при наложении периодических граничных условий перемещения на противоположных границах рассматриваемой области должны в обязательном порядке быть равны друг другу. При наложении контактных ограничений относительные перемещения узлов на контактирующих поверхностях либо полностью блокируются (bonded contact), либо допускаются лишь тангенциальные перемещения с дополнительной реакцией (frictional contact) или вовсе без нее (frictionless contact). 
Способ задания таких ограничений в зарубежной литературе носит Multiple Point Constraint (MPC), в русскоязычной литературе можно встретить название многоточечные связи. В данном пособии мы будем придерживаться международного обозначения и называть их просто MPC связями.

Существует несколько способов наложения MPC связей: метод штрафной функции (penalty function), метод исключения зависимых степеней свободы (master-slave illumination) и метод множителей Лагранжа. Мы ограничимся рассмотрением метода штрафной функции, как наиболее простого и в то же время, довольно универсального метода.

Рассмотрим стержневую систему, состоящую из 6 элементов и 7 узлов соответственно (см. рис.1). Для простоты положим, что данная система работает только лишь на растяжение и, соответственно, ее узлы имеют лишь по одной степени свободы.
Наложим условие равенства перемещений 2-го и 6-го узлов:
$$u_2=u_6$$
Для этого представим, что узлы 2 и 6 соединены жестким стержнем с осевой жесткостью $w$ многократно превышающей жесткость исходного стержня (см. рисунок 2). Обозначим этот дополнительный элемент индексом 7, как показано на рисунке 2.

Этот стержень называется штрафным элементом, а его жесткость $w$ - штрафным весом. Данный элемент является фиктивным, но по своей сути ничем не отличается от остальных стержневых элементов. Если продолжать сборку уравнений жесткости, то уравнение жесткости штрафного элемента будут иметь вид

$$
w
\begin{bmatrix}
\hspace{0.3cm}1 & -1\\
-1 & \hspace{0.3cm}1\\
\end{bmatrix}
\begin{Bmatrix}
u_2\\ u_6\\
\end{Bmatrix}
=
\begin{Bmatrix}
0\\
0\\
\end{Bmatrix}
$$

Поскольку на каждый узел приходится по одной степени свободы, две локальные степени свободы элемента отображаются в глобальные степени свободы узлов 2 и 6 соответственно. Используя правила сборки глобальной матрицы жесткости, можем записать новую модифицированную систему уравнений $\hat{K}\hat{u} = \hat{f}$

\begin{equation*}
\begin{bmatrix}
K_{11} &     K_{12} &      0 &      0 &      0 &          0 &      0 \\
K_{21} & K_{22} + w & K_{23} &      0 &0 & -w\hspace{0.3cm} &      0 \\
     0 &     K_{32} & K_{33} & K_{34} &      0 &          0 &      0 \\
     0 &          0 & K_{43} & K_{44} & K_{45} &          0 &      0 \\
     0 &          0 &      0 & K_{54} & K_{55} &     K_{56} &      0 \\
0 &-w\hspace{0.3cm} &      0 &      0 & K_{65} & K_{66} + w & K_{67} \\
     0 &          0 &      0 &      0 &      0 &     K_{76} & K_{77} \\
\end{bmatrix}
\begin{Bmatrix}
u_1\\ u_2 \\ u_3 \\ u_4 \\ u_5 \\ u_6 \\ u_7
\end{Bmatrix} = 
\begin{Bmatrix}
f_1\\ f_2 \\ f_3 \\ f_4 \\ f_5 \\ f_6 \\ f_7
\end{Bmatrix}
\end{equation*}

решив которую найдем решение удовлетворяющее дополнительным ограничениям. Так же следует обратить внимание, что в новой системе изменилась лишь матрица жесткости, в то время как вектор неизвестных и вектор внешних сил (правая часть)
